```python
import pandas as pd
from csv import reader
import csv
import os
df = pd.read_csv("/Users/bartgisone/Downloads/com.github.yeriomin.workoutlog.csv")
df = df.filter(['userName','packageName','score','content'], axis=1)
#removes special characters
df['content'] = df['content'].str.replace(r'[^\w\s]+','',regex=True)
df.head()
```

```python
#changes 0-9 to string form
df['content'] = df['content'].str.replace('0','zero',regex=True)
df['content'] = df['content'].str.replace('1','one',regex=True)
df['content'] = df['content'].str.replace('2','two',regex=True)
df['content'] = df['content'].str.replace('3','three',regex=True)
df['content'] = df['content'].str.replace('4','four',regex=True)
df['content'] = df['content'].str.replace('5','five',regex=True)
df['content'] = df['content'].str.replace('6','six',regex=True)
df['content'] = df['content'].str.replace('7','seven',regex=True)
df['content'] = df['content'].str.replace('8','eight',regex=True)
df['content'] = df['content'].str.replace('9','nine',regex=True)
df.head()
```

```python
#removes double spaces
df['content'] = df['content'].str.replace('  ',' ',regex=True)
df.head()
```

```python
#all words in lowercase
df['content'] = df['content'].str.lower()
df.head()
```

```python
#creates a new dataframe for lemmatization
df_new = df
df_new.head()
```

```python
#Removes stopwords
from nltk.corpus import stopwords
all_stopwords = set(stopwords.words('english'))
df_new['content']=df_new['content'].apply(str)
df_new['content'] = df['content'].apply(lambda x: ' '.join([item for item in x.split() if item not in all_stopwords]))

df_new.head()    
```

```python
#tokenizes and lemmatizes the 'content' portion 
#outputs all reviews
import nltk
import re
from nltk.stem.wordnet import WordNetLemmatizer

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = WordNetLemmatizer()

df_new = df

df_new['content']=df_new['content'].apply(str)

def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]


df_new['content_lemmatized'] = df_new['content'].apply(lambda x: lemmatize_text(x))

df_new.head()
```

```python
#-------------------------------Task 3-----------------------------------------------/ 
#uses textblob for finding Review Sentiment
import textblob
from textblob import TextBlob
df_tb = df_new

df_tb['Polarity'] = df_tb['content'].apply(lambda x: TextBlob(x).sentiment)
df_tb = df_tb.filter(['packageName','content','Polarity'], axis=1)

df_tb.head()
```

```python
#users Vader for Review Sentiment

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
df_vd = df_new
df_vd['Polarity'] = df['content'].apply(analyzer.polarity_scores)

df_vd = df_vd.filter(['packageName','content','Polarity'], axis=1)
df_vd.head()
```

```python
# answering Question V.

#The polaritys from both textblob and vader are similar in magnitude. 
#That is to say, if Vader is overwhelmingly, positive, neutral or negative, the textblob will also be similarily as positive, negative or neutral.
#for example, the 5th review reads 'awesome app perfect' and it is highly positive in both textblob and vader.
#The main difference between textblob and vader is how it outputs this sentiment

#vader splits the sentiment into positive, negative or neutral with a score of 0.0 - 1.0
#vader then gives a compound score from -1.0 - 1.0
#textblob on the otherhand has a single scale from 0.0-1.0 
#textblob also contains a subjectivity score from 0.0-1.0

#overall I prefer Vader, as it shows the polarity in three different terms, and puts those three terms together to give a single score.
#textblob may have a subjectivity score, but for reviews like the 4th review ('app works personal trainer front') we get a neutral score of 1.0 in vader but an overall score of 0.0 in textblob.
#thus, random or very neutral reviews may skew the data in textblob.

```
